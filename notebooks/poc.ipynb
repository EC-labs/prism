```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
matplotlib.style.use("bmh")
```

```python
def sigmoid(x, alpha, beta): 
    return 1/(1 + np.exp(- alpha * (x + beta)))

x = np.linspace(-5, 5, 100)
plt.plot(x, sigmoid(x, 100, -0.9))
plt.plot(x, sigmoid(x, 1, 0))
plt.xlim(0, 1)
plt.show()
```
